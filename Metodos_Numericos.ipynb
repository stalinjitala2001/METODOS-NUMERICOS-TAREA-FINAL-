1. Use el método de Euler para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial.


```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, t0, y0, h, t_final):
    N = int((t_final - t0) / h)
    t_values = np.linspace(t0, t_final, N + 1)
    y_values = np.zeros(N + 1)
    y_values[0] = y0

    for i in range(N):
        y_values[i + 1] = y_values[i] + h * f(t_values[i], y_values[i])

    return t_values, y_values

def plot_solution(t_values, y_values, title):
    plt.plot(t_values, y_values, 'o-', label='Aproximación de Euler')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Ejercicio 1a: y' = te^{3t} - 2y, y(0) = 0, h = 0.5, 0 ≤ t ≤ 1
def f1a(t, y):
    return t * np.exp(3 * t) - 2 * y

t_values, y_values = euler_method(f1a, t0=0, y0=0, h=0.5, t_final=1)
plot_solution(t_values, y_values, "Método de Euler para y' = te^{3t} - 2y")

# Ejercicio 1b: y' = 1 + (t - y)^2, y(2) = 1, h = 0.5, 2 ≤ t ≤ 3
def f1b(t, y):
    return 1 + (t - y)**2

t_values, y_values = euler_method(f1b, t0=2, y0=1, h=0.5, t_final=3)
plot_solution(t_values, y_values, "Método de Euler para y' = 1 + (t - y)^2")

# Ejercicio 1c: y' = 1 + y/t, y(1) = 2, h = 0.25, 1 ≤ t ≤ 2
def f1c(t, y):
    return 1 + y / t

t_values, y_values = euler_method(f1c, t0=1, y0=2, h=0.25, t_final=2)
plot_solution(t_values, y_values, "Método de Euler para y' = 1 + y/t")

# Ejercicio 1d: y' = cos(2t) + sen(3t), y(0) = 1, h = 0.25, 0 ≤ t ≤ 1
def f1d(t, y):
    return np.cos(2 * t) + np.sin(3 * t)

t_values, y_values = euler_method(f1d, t0=0, y0=1, h=0.25, t_final=1)
plot_solution(t_values, y_values, "Método de Euler para y' = cos(2t) + sen(3t)")
```

aproximaciones del ejercicio 3.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Método de Euler
def euler_method(f, t0, y0, h, t_final):
    N = int((t_final - t0) / h)
    t_values = np.linspace(t0, t_final, N + 1)
    y_values = np.zeros(N + 1)
    y_values[0] = y0

    for i in range(N):
        y_values[i + 1] = y_values[i] + h * f(t_values[i], y_values[i])

    return t_values, y_values

# Ejercicio 4: Soluciones reales
def exact_solution_4a(t):
    return t / (1 + np.log(t))

def exact_solution_4b(t):
    return t * np.tan(np.log(t))

def exact_solution_4c(t):
    return -3 + 2 / (1 + np.exp(-2*t))

def exact_solution_4d(t):
    return t**2 + (1/3) * np.exp(-5*t)

# Funciones para los literales del ejercicio 3
def f3a(t, y):
    return y/t - (y/t)**2

def f3b(t, y):
    return 1 + y/t + (y/t)**2

def f3c(t, y):
    return -(y + 1) * (y + 3)

def f3d(t, y):
    return -5*y + 5*t**2 + 2*t

# Calcular errores para cada literal en ejercicio 4
# Literal 4a
t_values, y_values = euler_method(f3a, t0=1, y0=1, h=0.1, t_final=2)
exact_values_4a = exact_solution_4a(t_values)
error_4a = np.abs(y_values - exact_values_4a)

# Literal 4b
t_values_b, y_values_b = euler_method(f3b, t0=1, y0=0, h=0.2, t_final=3)
exact_values_4b = exact_solution_4b(t_values_b)
error_4b = np.abs(y_values_b - exact_values_4b)

# Literal 4c
t_values_c, y_values_c = euler_method(f3c, t0=0, y0=-2, h=0.2, t_final=2)
exact_values_4c = exact_solution_4c(t_values_c)
error_4c = np.abs(y_values_c - exact_values_4c)

# Literal 4d
t_values_d, y_values_d = euler_method(f3d, t0=0, y0=1/3, h=0.1, t_final=1)
exact_values_4d = exact_solution_4d(t_values_d)
error_4d = np.abs(y_values_d - exact_values_4d)

# Graficar errores
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(t_values, error_4a, 'o-', label='Error real para 4a')
plt.xlabel('t')
plt.ylabel('Error')
plt.title('Error real en 4a')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t_values_b, error_4b, 'o-', label='Error real para 4b')
plt.xlabel('t')
plt.ylabel('Error')
plt.title('Error real en 4b')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(t_values_c, error_4c, 'o-', label='Error real para 4c')
plt.xlabel('t')
plt.ylabel('Error')
plt.title('Error real en 4c')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t_values_d, error_4d, 'o-', label='Error real para 4d')
plt.xlabel('t')
plt.ylabel('Error')
plt.title('Error real en 4d')
plt.grid(True)

plt.tight_layout()
plt.show()


```

Use el método de Taylor de orden 2 para aproximar las soluciones para cada uno de los siguientes problemas de
valor inicial.

```python
import numpy as np

# Función para resolver con método de Taylor de orden 2
def taylor_orden2(f, df_t, df_y, t0, y0, h, n_steps):
    t_values = [t0]
    y_values = [y0]

    for i in range(n_steps):
        t = t_values[-1]
        y = y_values[-1]

        y_next = y + h * f(t, y) + (h**2 / 2) * (df_t(t, y) + df_y(t, y) * f(t, y))
        t_next = t + h

        t_values.append(t_next)
        y_values.append(y_next)

    return np.array(t_values), np.array(y_values)

# Definimos las funciones dadas y sus derivadas parciales

# a) y' = t e^(3t) - 2y
f_a = lambda t, y: t * np.exp(3*t) - 2*y
df_t_a = lambda t, y: np.exp(3*t) + 3*t * np.exp(3*t)
df_y_a = lambda t, y: -2

# b) y' = 1 + (t - y)^2
f_b = lambda t, y: 1 + (t - y)**2
df_t_b = lambda t, y: 2 * (t - y)
df_y_b = lambda t, y: -2 * (t - y)

# c) y' = 1 + y/t
f_c = lambda t, y: 1 + y/t
df_t_c = lambda t, y: -y/t**2
df_y_c = lambda t, y: 1/t

# d) y' = cos(2t) + sin(3t)
f_d = lambda t, y: np.cos(2*t) + np.sin(3*t)
df_t_d = lambda t, y: -2*np.sin(2*t) + 3*np.cos(3*t)
df_y_d = lambda t, y: 0

# Inicialización de parámetros
t0_a, y0_a, h_a, n_steps_a = 0, 0, 0.5, 2
t0_b, y0_b, h_b, n_steps_b = 2, 1, 0.5, 2
t0_c, y0_c, h_c, n_steps_c = 1, 2, 0.25, 4
t0_d, y0_d, h_d, n_steps_d = 0, 1, 0.25, 4

# Soluciones con el método de Taylor de orden 2
t_values_a, y_values_a = taylor_orden2(f_a, df_t_a, df_y_a, t0_a, y0_a, h_a, n_steps_a)
t_values_b, y_values_b = taylor_orden2(f_b, df_t_b, df_y_b, t0_b, y0_b, h_b, n_steps_b)
t_values_c, y_values_c = taylor_orden2(f_c, df_t_c, df_y_c, t0_c, y0_c, h_c, n_steps_c)
t_values_d, y_values_d = taylor_orden2(f_d, df_t_d, df_y_d, t0_d, y0_d, h_d, n_steps_d)

print("Taylor Orden 2:")
print("a)", list(zip(t_values_a, y_values_a)))
print("b)", list(zip(t_values_b, y_values_b)))
print("c)", list(zip(t_values_c, y_values_c)))
print("d)", list(zip(t_values_d, y_values_d)))
```

```python
import numpy as np

# Función para resolver con método de Taylor de orden 4
def taylor_orden4(f, df_t, d2f_tt, d3f_ttt, t0, y0, h, n_steps):
    t_values = [t0]
    y_values = [y0]

    for i in range(n_steps):
        t = t_values[-1]
        y = y_values[-1]

        y_next = y + h * f(t, y) + (h**2 / 2) * df_t(t, y) + (h**3 / 6) * d2f_tt(t, y) + (h**4 / 24) * d3f_ttt(t, y)
        t_next = t + h

        t_values.append(t_next)
        y_values.append(y_next)

    return np.array(t_values), np.array(y_values)

# Aquí se definen las segundas y terceras derivadas para cada función

# Para la función a)
d2f_tt_a = lambda t, y: 9*t*np.exp(3*t) + 6*np.exp(3*t)
d3f_ttt_a = lambda t, y: 27*t*np.exp(3*t) + 27*np.exp(3*t)

# Para la función b)
d2f_tt_b = lambda t, y: 2
d3f_ttt_b = lambda t, y: 0

# Para la función c)
d2f_tt_c = lambda t, y: 2*y/t**3
d3f_ttt_c = lambda t, y: -6*y/t**4

# Para la función d)
d2f_tt_d = lambda t, y: -4*np.cos(2*t) - 9*np.sin(3*t)
d3f_ttt_d = lambda t, y: 8*np.sin(2*t) - 27*np.cos(3*t)

# Inicialización de parámetros
t0_a, y0_a, h_a, n_steps_a = 0, 0, 0.5, 2
t0_b, y0_b, h_b, n_steps_b = 2, 1, 0.5, 2
t0_c, y0_c, h_c, n_steps_c = 1, 2, 0.25, 4
t0_d, y0_d, h_d, n_steps_d = 0, 1, 0.25, 4

# Soluciones con el método de Taylor de orden 4
t_values_a, y_values_a = taylor_orden4(f_a, df_t_a, d2f_tt_a, d3f_ttt_a, t0_a, y0_a, h_a, n_steps_a)
t_values_b, y_values_b = taylor_orden4(f_b, df_t_b, d2f_tt_b, d3f_ttt_b, t0_b, y0_b, h_b, n_steps_b)
t_values_c, y_values_c = taylor_orden4(f_c, df_t_c, d2f_tt_c, d3f_ttt_c, t0_c, y0_c, h_c, n_steps_c)
t_values_d, y_values_d = taylor_orden4(f_d, df_t_d, d2f_tt_d, d3f_ttt_d, t0_d, y0_d, h_d, n_steps_d)

print("Taylor Orden 4:")
print("a)", list(zip(t_values_a, y_values_a)))
print("b)", list(zip(t_values_b, y_values_b)))
print("c)", list(zip(t_values_c, y_values_c)))
print("d)", list(zip(t_values_d, y_values_d)))
```
